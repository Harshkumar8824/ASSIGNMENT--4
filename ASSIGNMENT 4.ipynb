```python
import pandas as pd
```

```python
csv_file = "student-mat.csv"
```

```python
data = pd.read_csv(csv_file)
```

```python
binary_attributes = ["failures", "absences"]  
numerical_attributes = ["age", "famsize"] 
nominal_attributes = ["guardian", "sex"] 
binary_matrix = data[binary_attributes].values
numerical_matrix = data[numerical_attributes].values
nominal_matrix = pd.get_dummies(data[nominal_attributes]).values 
```

```python
print("Binary Matrix:")
print(binary_matrix)
```

```python
print("Numerical Matrix:")
print(numerical_matrix)
```

```python
print("Nominal Matrix:")
print(nominal_matrix)
```
